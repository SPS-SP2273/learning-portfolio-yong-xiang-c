```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### pyplot Interface

```python
plt.plot(x, sin_x, label='$\\sin x$')
plt.plot(x, cos_x, label='$\\cos x$')
plt.legend()
```

### OO Interface

```python
fig, ax = plt.subplots()

plt.plot(x, sin_x, label='$\\sin x$')
plt.plot(x, cos_x, label='$\\cos x$')
plt.legend()
```

```python
fig, ax = plt.subplots(
    nrows=2,                    # ncols changes the number of columns and puts them side by side
    figsize=(8, 4)
)

my_sin = ax[0]
my_cos = ax[1]        # or my_sin, my_cos = ax[0], ax[1]     or my_sin, my_cos = ax

ax[0].plot(x, sin_x, label='$\\sin x$')
# my_sin.legend()

ax[1].plot(x, cos_x, label='$\\cos x$')
# my_cos.legend()

my_sin.fill_between(x,0,sin_x, alpha=.5, color='orange')     # alpha is to change the opacity of the fill.
my_cos.fill_between(x,0,cos_x, alpha=.5, color='yellow')

for a in [my_sin, my_cos]:
    a.legend()
    a.set_xlim([-np.pi,np.pi])

plt.suptitle('Trig Functions')

my_sin.set_xlabel('Angle')
my_cos.set_xlabel('Angle')
my_sin.set_ylabel('$\\sin x$', rotation=0)
my_cos.set_ylabel('$\\cos x$', rotation=0)

plt.tight_layout()

plt.savefig('Woohoo.png',dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(
    nrows=2,ncols=2,
    figsize=(8, 4)
)

for row in [0,1]:
    for col in [0,1]:
        ax[row,col].plot(x, cos_x)

for a in ax.flatten():
    a.grid(color='black')
    a.grid(alpha=.3)
```
