## Exercise 3 (Timing Stuff) ##

```python
squares = [i**2 for i in range(10)]
print(squares)
```

```python
squares = list()
for i in range(10):
    squares.append(i**2)
print(squares)
```

Both codes give the same result of [0, 1, 4, 9, 16, 25, 36, 49, 64, 81].

```python
%timeit squares = [i**2 for i in range(10)]
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
```

The %%timeit output represents the time taken for the entire code to run. As the first code took a shorter time as compared to the second code, the first code would be more efficient (considering that they give the same output), hence the first code is preferred.
