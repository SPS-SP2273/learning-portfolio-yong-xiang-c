```python
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
def my_line_py(m,c):
    x_values = [x for x in range(100)]
    y_values = [m+x_value + c + 50*random.random() for x_value in x_values]
    return x_values, y_values
```

```python
for m in range(5):
    x, y = my_line_py(m,10)
    plt.plot(x,y)
```
